```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import numpy as np

# Load your raw data
csv_file_path = "Desktop/Dragonfly sesh/VolumeSurfaceLindblad-Theta-Hb.csv"
df = pd.read_csv(csv_file_path)

# Ensure your data columns are named correctly
x = df['VolumeSA_Lindblad 2005(mm)']
y = df['Theta']

# Create 2D histogram with log-scaled color normalization
plt.figure(figsize=(10, 8))
plt.hist2d(
    x, y, bins=[75, 75],
    cmap='viridis', # color options: cividis, plasma, inferno, magma, viridis, Spectral, turbo
    norm=LogNorm()  # Fixed line to log transform values of both axes
)

plt.colorbar(label='Count')
plt.xscale('log') #log X-axis
#plt.yscale('log') #log Y-axis

plt.xlabel('Volume/Surface Area (Lindblad 2005)')
plt.ylabel('Theta')
plt.title('Heatmap of Mitochondrial Volume/Surface Area over Theta')

# Set X-axis ticks manually or sparsely
#ax = plt.gca()
#xticks = ax.get_xticks()
#ax.set_xticks(xticks[::5])  # Show every 5th tick

plt.tight_layout()
plt.savefig("Desktop/Dragonfly sesh/VolumeSA-theta.png", dpi=300, bbox_inches='tight') #activate this line when you are ready to save

plt.show()
```
