```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import numpy as np

# Load your raw data
csv_file_path = "Desktop/Dragonfly sesh/Sphercity-AspectRatio-Hb.csv"
df = pd.read_csv(csv_file_path)

# Ensure your data columns are named correctly
x = df['AspectRatio']
y = df['Sphericity']

# Create 2D histogram with log-scaled color normalization
plt.figure(figsize=(10, 8))
plt.hist2d(
    x, y, bins=[75, 75],
    cmap='viridis', # color options: cividis, plasma, inferno, magma, viridis, Spectral, turbo
    norm=LogNorm()  # Fixed line to log transform values of both axes
)

plt.colorbar(label='Count')
plt.xscale('log') #log X-axis
plt.yscale('log') #log Y-axis

plt.xlabel('Aspect Ratio')
plt.ylabel('Sphericity')
plt.title('Log-Scaled Heatmap of Mitochondrial Sphericity over Aspect Ratio')

# Set X-axis ticks manually or sparsely
#ax = plt.gca()
#xticks = ax.get_xticks()
#ax.set_xticks(xticks[::5])  # Show every 5th tick

plt.tight_layout()
plt.savefig("Desktop/Dragonfly sesh/sphericity-aspectratio.png", dpi=300, bbox_inches='tight') #activate this line when you are ready to save

plt.show()
```
